# Basic Open-CV

```python
import cv2 as cv
import sys
```

```python
img = cv.imread("../img/suscy.png", cv.IMREAD_GRAYSCALE)
if img is None:
    sys.exit("Could not read the image.")
cv.imshow("Display window", img)
k = cv.waitKey(0)
if k == ord("s"):
    cv.imwrite("../img/suscy_gray.png", img)
    cv.destroyAllWindows()
elif k == ord("q"):
    cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.imshow(img, cmap = "gray", interpolation = "bicubic")
plt.xticks([]), plt.yticks([])
plt.show()
```
